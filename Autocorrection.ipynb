```python
import nltk
from nltk.corpus import words, stopwords
from collections import Counter
nltk.download('words')
nltk.download('stopwords')
word_list = set(words.words())  
stop_words = set(stopwords.words('english'))
def levenshtein_distance_dp(str1, str2):
    m = len(str1)
    n = len(str2)

    dp = [[0] * (n + 1) for _ in range(m + 1)]

    for i in range(m + 1):
        for j in range(n + 1):
            if i == 0:
                dp[i][j] = j
            elif j == 0:
                dp[i][j] = i
            elif str1[i - 1] == str2[j - 1]:
                dp[i][j] = dp[i - 1][j - 1]
            else:
                dp[i][j] = min(dp[i - 1][j] + 1, dp[i][j - 1] + 1, dp[i - 1][j - 1] + 1)

    return dp[m][n]

def correct_spelling_with_common_words(word, word_list):
    if word in stop_words:  
        return word
    min_distance = float('inf')
    corrected_words = []
    for w in word_list:
        if abs(len(word) - len(w)) <= 2: 
            distance = levenshtein_distance_dp(word, w)
            if distance < min_distance:
                min_distance = distance
                corrected_words = [w]
            elif distance == min_distance:
                corrected_words.append(w)

    return corrected_words
wrong_word = input("Səhv yazılmış sözü daxil edin: ")

corrected = correct_spelling_with_common_words(wrong_word, word_list)

if isinstance(corrected, list):
    print(f"Original Word: {wrong_word}")
    top_3 = sorted(corrected, key=lambda x: levenshtein_distance_dp(wrong_word, x))[:3]
    print(f"Top 3 Corrected Words: {', '.join(top_3)}")
else:
    print(f"Original Word: {wrong_word}")
    print(f"Word is a stopword, no correction needed.")
```
